```python
# Filter the DataFrame to include only necessary columns
df_filtered = df[['HOTEL', 'ARRIVAL_DATE_MONTH', 'IS_CANCELED']]

# Group by 'HOTEL' and 'ARRIVAL_DATE_MONTH' to calculate the count of bookings and cancellations
grouped_df = df_filtered.groupby(['HOTEL', 'ARRIVAL_DATE_MONTH']).sum().reset_index()

# Pivot the DataFrame to have 'ARRIVAL_DATE_MONTH' as index and 'HOTEL' as columns
pivot_df = grouped_df.pivot(index='ARRIVAL_DATE_MONTH', columns='HOTEL', values='IS_CANCELED')

# Plot the line plot
plt.figure(figsize=(10, 6))
pivot_df.plot(kind='line', marker='o')
plt.title('Booking Status by Month and Hotel')
plt.xlabel('Month')
plt.ylabel('Number of Bookings/Cancellations')
plt.grid(True)
plt.legend(title='Hotel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total length of stay (weekend nights + week nights)
df['TOTAL_LENGTH_OF_STAY'] = df['STAYS_IN_WEEKEND_NIGHTS'] + df['STAYS_IN_WEEK_NIGHTS']

# Define bins for length of stay categories
bins = [0, 3, 7, 14, 30, float('inf')]
labels = ['Short (1-3 days)', 'Medium (4-7 days)', 'Long (8-14 days)', 'Extended (15-30 days)', 'Very Extended (>30 days)']

# Create a new column 'LENGTH_OF_STAY_CATEGORY' to categorize length of stay
df['LENGTH_OF_STAY_CATEGORY'] = pd.cut(df['TOTAL_LENGTH_OF_STAY'], bins=bins, labels=labels)

# Plot pie plot for length of stay categories
plt.figure(figsize=(8,8 ))
df['LENGTH_OF_STAY_CATEGORY'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Length of Stay Categories')
plt.ylabel('')
plt.show()
```

This code creates a new column TOTAL_LENGTH_OF_STAY by summing up the number of stays in weekend nights (STAYS_IN_WEEKEND_NIGHTS) and the number of stays in week nights (STAYS_IN_WEEK_NIGHTS). Then it defines bins and labels for different categories of length of stay, categorizes the length of stay into these categories using pd.cut(), and stores the result in a new column LENGTH_OF_STAY_CATEGORY. Finally, it plots a pie chart to visualize the distribution of length of stay categories.

Here's a breakdown of the code:

df['TOTAL_LENGTH_OF_STAY'] = df['STAYS_IN_WEEKEND_NIGHTS'] + df['STAYS_IN_WEEK_NIGHTS']: Calculates the total length of stay by adding the number of stays in weekend nights and week nights.

```python
# Group by 'ARRIVAL_DATE_YEAR' and 'HOTEL' to calculate the number of cancellations for each combination of year and hotel
cancellations_by_year_and_hotel = df.groupby(['ARRIVAL_DATE_YEAR', 'HOTEL'])['IS_CANCELED'].sum()

# Plot bar plot for cancellation analysis by year and hotel
plt.figure(figsize=(12, 6))
cancellations_by_year_and_hotel.unstack().plot(kind='bar', stacked=True)
plt.title('Cancellation Analysis by Year and Hotel')
plt.xlabel('Year')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=45)
plt.legend(title='Hotel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
